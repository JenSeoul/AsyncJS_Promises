## 1. Ï£ºÏñ¥ÏßÑ Î¶¨Ïä§Ìä∏Ïóê ÎåÄÌï¥ÏÑú ÌèâÍ∑†, Î∂ÑÏÇ∞, ÌëúÏ§ÄÌé∏Ï∞®Î•º Íµ¨ÌïòÎäî Ìï®ÏàòÎ•º ÏûëÏÑ±Ìï¥Î≥¥ÏÑ∏Ïöî. (`mean`Í∞ôÏùÄ builtin function Ï†úÏô∏)
> Ïù¥ÌõÑ Îã§Ïùå Îç∞Ïù¥ÌÑ∞ [243, 278, 184, 249, 207] Ïóê ÎåÄÌï¥ÏÑú Ìï¥Îãπ Ìï®ÏàòÎ•º ÏÇ¨Ïö©Ìïú Í≤∞Í≥ºÎ•º Î≥¥Ïù¥ÏÑ∏Ïöî.

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt

v = np.array([243,278,184,249,207])
v2 = np.array([88,89,83,112,104])


# ÏûëÏÑ±ÌïòÏÑ∏Ïöî
def mymean(v):
  v = np.array(v)
  return sum(v)/len(v)

# ÏûëÏÑ±ÌïòÏÑ∏Ïöî.
def myvar(v):
  mean = mymean(v)
  variance = np.square(v - mean).sum()/len(v)
  return variance

# ÏûëÏÑ±ÌïòÏÑ∏Ïöî.
def mystd(v):
  variance = myvar(v)
  return np.sqrt(variance)
```

```python
res1 = mymean(v)
res2 = myvar(v)
res3 = mystd(v)
```

## 2. ÏúÑÏóêÏÑú ÏÇ¨Ïö©Îêú Îç∞Ïù¥ÌÑ∞ÏôÄ Îã§Ïùå Îç∞Ïù¥ÌÑ∞Ïùò Í≥µÎ∂ÑÏÇ∞Í≥º, ÏÉÅÍ¥ÄÍ≥ÑÏàòÎ•º Íµ¨Ìï¥Î≥¥ÏÑ∏Ïöî [88, 89, 83, 112, 104] 

```python
#cov function 
def mycov(x,y):
  meanx = mymean(x)
  meany = mymean(y)
  sum = 0
  for i in range(0, len(x)):
        sum += ((x[i] - meanx) * (y[i] - meany))
  return sum/(len(x)-1)

#corr function 
def mycorr(x,y):
  cov = round(mycov(x,y),1)
  xs = np.std(x,ddof=1)
  ys = np.std(y,ddof=1)
  r = cov/(xs*ys)
  return r
```

```python
#answer 
res4 = mycov(v,v2)
res5 = mycorr(v,v2)
```

- (x, y) Îäî (0, 0) ÏóêÏÑú (x, y)Î°ú Í∞ÄÎäî Î≤°ÌÑ∞Îùº Í∞ÄÏ†ïÌï©ÎãàÎã§.

Ïù¥ÌõÑ

- ÏûÖÎ†•Îêú Îç∞Ïù¥ÌÑ∞Î•º ÌååÎûÄÏÉâ ÏÑ†ÏúºÎ°ú, 
- `y = x` ÎùºÎäî Î≤°ÌÑ∞Î•º Îπ®Í∞ÑÏÉâ ÏÑ†ÏúºÎ°ú, ÎßàÏßÄÎßâÏúºÎ°ú 
- projection Îêú ÏÑ†ÏùÑ ÎÖπÏÉâ Ï†êÏÑ†(dashed)ÏúºÎ°ú Í∑∏ÎûòÌîÑÏóê Í∑∏Î¶¨ÏÑ∏Ïöî.

```python
v = [7, 4]
w = np.array([7,7])

# Ï±ÑÏö∞ÏÑ∏Ïöî.
def myProjection(v):
  v = np.array(v)
  w_norm = np.linalg.norm(w)    
  vprime = (np.dot(v, w) / np.dot(w, w)) * w
  return vprime

vprime = myProjection(v)
```

```python
plt.figure(figsize = (7,7))
plt.xlim(0,7)          
plt.ylim(0,7)
plt.xticks(ticks=np.arange(0, 7, step=1))
plt.yticks(ticks=np.arange(0, 7, step=1))

vprime = myProjection(v)
x_minus_proj = vprime[0]-v[0]
y_minus_proj = vprime[1]-v[1]

axes = plt.gca()
plt.gca().set_aspect('equal')
x_vals = np.array(axes.get_xlim())
y_vals = x_vals

plt.arrow(0, 0, vprime[0], vprime[1],linewidth=5, linestyle = '--', color = 'green',alpha=0.8)
plt.arrow(0, 0, v[0], v[1], linewidth = 3, head_width = .05, head_length = .05, color = 'b')
plt.arrow(v[0], v[1],x_minus_proj, y_minus_proj, linewidth = 3, head_width = .05, head_length = .05, color = 'orange')
plt.arrow(0, 0,7, 7, linewidth = 3, head_width = .05, head_length = .05, color = 'red',alpha=0.5)

plt.grid()
plt.show()
```
